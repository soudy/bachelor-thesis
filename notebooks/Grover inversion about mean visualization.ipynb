```python
import matplotlib.pyplot as plt
import numpy as np


plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'''
\usepackage{palatino}
\usepackage{newpxmath}''')

N = 2**2
COLORS = ['#1f77b4', '#ff7f0e', '#2ca02c']

fig, ax = plt.subplots(figsize=(4, 2), dpi=190)

states = [r'$|00\rangle$', r'$|01\rangle$', r'$|10\rangle$', r'$|11\rangle$']
probs = [1/np.sqrt(N), 1/np.sqrt(N), 1/np.sqrt(N), 1/np.sqrt(N)]
ax.bar(states, probs, 0.02, color='black')

ax.set_yticks(np.arange(-1, 1.2, 0.5))
ax.axhline(linewidth=1, color='black')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.ylabel(r'$\textnormal{Re}(\alpha_i)$')
plt.savefig('figures/aa_initial_equal.pdf', bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 2), dpi=190)

states = [r'$|00\rangle$', r'$|01\rangle$', r'$|10\rangle$', r'$|11\rangle$']
probs = [1/np.sqrt(N), 1/np.sqrt(N), -1/np.sqrt(N), 1/np.sqrt(N)]
mean = np.mean(probs)
ax.bar(states, probs, 0.02, color='black')

ax.set_yticks(np.arange(-1, 1.2, 0.5))
ax.axhline(linewidth=1, color='black')
ax.axhline(linewidth=1, linestyle='--', color=COLORS[1], y=mean)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.ylabel(r'$\textnormal{Re}(\alpha_i)$')
plt.savefig('figures/aa_phase_flipped_mean.pdf', bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 2), dpi=190)

states = [r'$|00\rangle$', r'$|01\rangle$', r'$|10\rangle$', r'$|11\rangle$']
probs = np.array([1/np.sqrt(N), 1/np.sqrt(N), -1/np.sqrt(N), 1/np.sqrt(N)])
mean = np.mean(probs)
inverted_probs = mean + (mean - probs)
ax.bar(states, inverted_probs, 0.02, color='black')

ax.set_yticks(np.arange(-1, 1.2, 0.5))
ax.axhline(linewidth=1, color='black')
ax.axhline(linewidth=1, linestyle='--', color=COLORS[1], y=mean)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.ylabel(r'$\textnormal{Re}(\alpha_i)$')
plt.savefig('figures/aa_diffused.pdf', bbox_inches='tight')
plt.show()
```
