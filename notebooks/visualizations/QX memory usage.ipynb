```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

mem = np.load("./data/mem_usage_qx.npz", allow_pickle=True)
df = pd.DataFrame.from_dict(mem["data"].item(), orient="index")
df["mean"] = df.mean(axis=1)
df["std"] = df.std(axis=1)
df
```

```python
import humanize

for row in df.itertuples():
    print(f"{row.Index} & $\\text{{{humanize.naturalsize(row.mean, format='%.2f')}}} \pm \\text{{{humanize.naturalsize(row.std, format='%.2f')}}}$ \\\\")
```

```python
from scipy.optimize import curve_fit

x = np.array(df.index)
y = df["mean"]/1000/1000/1000

def exp_func(x, a, b, c):
    return a * np.exp(b * x) + c

popt, pcov = curve_fit(exp_func, x, y+df["std"]/1000/1000/1000)
popt
```

```python
residuals = y - exp_func(df.index, *popt)
ss_res = np.sum(residuals**2)
ss_tot = np.sum((y - np.mean(y))**2)
r_squared = 1 - (ss_res / ss_tot)
r_squared
```

```python
plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'''
\usepackage{palatino}
\usepackage{newpxmath}''')
plt.rc('font', family='serif')

fig, ax = plt.subplots(figsize=(4.35, 3.35), dpi=300)
plt.yscale('log')

ax.scatter(df.index, y, marker="x", s=20,  color="black", label="measured", zorder=1, linewidth=2)
ax.plot(df.index, exp_func(df.index, *popt), color="orange", label=r"$\hat{m}(n)$", linewidth=1, alpha=0.85, zorder=2)

plt.legend(fontsize=9)
plt.ylabel('RSS memory usage in GB')
plt.xlabel('$n$')
plt.savefig('figures/qx_memory_plot.pdf', bbox_inches='tight')
plt.show()
```
