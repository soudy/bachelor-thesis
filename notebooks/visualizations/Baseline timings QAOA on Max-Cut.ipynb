```python
import numpy as np
import pandas as pd

dfs = {}

for workflow in ("hpc", "qpu", "local"):
    dfs[workflow] = None
    
    for i in range(1, 6):
        data = np.load(f"data/qaoa_baseline/{workflow}/{i}.npz", allow_pickle=True)
        df = pd.DataFrame.from_records(data["optimization_steps"])
        
        if dfs[workflow] is None:
            dfs[workflow] = df
        else:
            dfs[workflow] = dfs[workflow].append(df)
```
