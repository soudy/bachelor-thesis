```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'''
\usepackage{palatino}
\usepackage{newpxmath}''')

a = '#bdc3c7'
b = '#000000'
cut_edge = '#e74c3c'

n = 5
V = np.arange(0, n, 1)
E =[(0, 1, 1.0), (0, 4, 1.0), (1, 2, 1.0), (4, 3, 1.0), (2, 3, 1.0), (1, 4, 1.0)] 

G = nx.Graph()
G.add_nodes_from(V)
G.add_weighted_edges_from(E)

edge_labels = nx.get_edge_attributes(G, 'weight')
pos = nx.spring_layout(G, seed=33)
pos = nx.spectral_layout(G, scale=2)

plt.subplots(figsize=(3.25, 2.25), dpi=190)
default_axes = plt.axes(frameon=False)

nx.draw_networkx(G, node_size=500, alpha=1, ax=default_axes, pos=pos,
                 node_color=[a, a, b, a, b],
                 edge_color=['k', cut_edge, cut_edge, cut_edge, cut_edge, cut_edge],
                 width=3,
                 font_color='#ffffff'
                )

plt.axis('off')
axis = plt.gca()
axis.set_xlim([1.1*x for x in axis.get_xlim()])
axis.set_ylim([1.15*y for y in axis.get_ylim()])
plt.tight_layout()
plt.savefig('figures/maxcut_5_graph_cut_example.pdf', bbox_inches='tight')
plt.show()
```
